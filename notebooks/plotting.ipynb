```python
class FiatUno:
    def __init__(self, color="white", radio=False):
        self.color = color
        self.radio = radio
        self.radio_is_on = False
    def turn_on_radio(self):
        self.radio_is_on = True
        print(f"Radio is on!")
    def turn_off_radio(self, a, b):
        self.radio_is_on = False
        print(f"Radio is off!")
```
